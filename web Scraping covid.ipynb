```python
#Using the beautiful soup library, print out the text of the h1 tag in the page.
from bs4 import BeautifulSoup
import requests
url =BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup = requests.get(url)
soup = soup.text
```

```python
code = BeautifulSoup(soup, "lxml")
code =code.table
```

```python
data = []
tags = code.find_all('tr')
for tag in tags:
    x = tag.text
    data.append(x)
data
```

```python
datas = []
for i in data:
    datas.append(i.split('\n')[1:-1])
datas
```

```python
clean_data = [i for i in datas if i[0] != '']
clean_data
```

```python
import csv
try:
    file = open('covid_data.csv', 'w')
    x = csv.writer(file)
    for i in clean_data:
        x.writerow(i)
    file.close()
except:
    print('File already exists')
```

```python
import pandas as pd 
df = pd.read_csv('covid_data.csv', encoding='ISO-8859-1')
df 
```

```python
country = list (df['Country,Other']) [0:10]
total_cases = list (df['TotalCases']) [0:10]
total_cases
cleaned_case = []
for case in total_cases:
    x = int(case.replace(',',''))
    cleaned_case.append(x)

total_cases
cleaned_case
```
